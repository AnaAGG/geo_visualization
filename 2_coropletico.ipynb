```python
# we loop through the dictionary to obtain the name of the communities in the json file
comunidades = []
for index in range(len(geo_json['features'])):
    comunidades.append(geo_json['features'][index]['properties']['name']) 
comunidades
```

```python
import random 
diccionario = {x : random.randint(0,1000) for x in comunidades}
```

```python
df = pd.DataFrame.from_dict(diccionario, orient='index').reset_index()
columnas=['comunidad', 'count']
df.columns = columnas
df.head()
```

```python
import folium

# create a plain world map
communities_map = folium.Map(location=[40.416775, -3.703790], zoom_start=6) # <-- ,tiles='stamenwatercolor'

# generate choropleth map 
communities_map.choropleth(
    geo_data=geo_json,
    data=df,
    columns=['comunidad', 'count'],
    key_on='feature.properties.name',
    fill_color='YlGnBu', 
    fill_opacity=1, 
    line_opacity=1,
    legend_name='Datitos inventados por comunidad',
    smooth_factor=0)

# display map
communities_map
```
